```python
import numpy as np

def descent_down_parabola (w_start, learning_rate, num_steps):
  result = [w_start, ]
  w_old = w_start
  for _ in range(num_steps):
    w_new = w_old - learning_rate * 2 * w_old
    result.append(w_new)
    w_old = w_new

  return result
```

```python
print(descent_down_parabola(10, 0.3, 5))
print(descent_down_parabola(-10, 0.3, 5))
```

```python
import matplotlib.pyplot as plt
# 함수값과 0의 차이에 대한 그래프 그리기
res_array = descent_down_parabola(10, 0.3, 100)
x = np.arange(101)
plt.plot(x, res_array)
plt.yscale("log")
plt.xlabel('Number of Steps')
plt.ylabel('Disatance from Minimum')
plt.title('Analyzing Convergence of Gradient Descent on $\mathscr{L}(w) = w^2$')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
# 함수값과 0의 차이에 대한 그래프 그리기
res_array = descent_down_parabola(10, 3, 100)
res_array = np.abs(res_array)
x = np.arange(101)
plt.plot(x, res_array)
plt.yscale("log")
plt.xlabel('Number of Steps')
plt.ylabel('Disatance from Minimum')
plt.title('Analyzing Convergence of Gradient Descent on $\mathscr{L}(w) = w^2$')
plt.grid()
plt.show()
```

```python
import numpy as np
def descent_down_2d_parabola (w_start, learning_rate, num_steps):
  result = [w_start]
  w_old = w_start
  for _ in range(num_steps):
    w_new = w_old - learning_rate * (np.array([4., 6.]) * w_old)
    result.append(w_new)
    w_old = w_new

  return result
```

```python
result = descent_down_2d_parabola([2.0, 4.0], 0.1, 100)
print(result)
```

```python
import matplotlib.pyplot as plt
x = np.arange(101)
plt.plot(x, result)
plt.xlabel('Number of Steps')
plt.ylabel('Disatance from Minimum')
plt.yscale("log")
plt.title('Analyzing Convergence of Gradient Descent on $\mathscr{L}(w) = w^2$')
plt.grid()
plt.show()
```
